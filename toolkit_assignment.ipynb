
1. What is NumPy, and why is it widely used in Python?

Why is it widely used?

Faster computations due to optimized C implementation

Supports vectorized operations

2. How does broadcasting work in NumPy?
Broadcasting allows operations between arrays of different shapes without explicit looping. NumPy automatically expands smaller arrays to match the dimensions of larger arrays.
Example:

python
Copy
Edit
import numpy as np
a = np.array([[1, 2, 3], [4, 5, 6]])
b = np.array([1, 2, 3])
print(a + b)  # b is broadcasted across each row of a

17. Explain the difference between apply() and map() in Pandas.
apply() works on DataFrame rows/columns and allows custom functions.

map() works on Series and is used for element-wise transformations.

Example:

python
Copy
Edit
df['Column'].apply(lambda x: x * 2)  # Apply function to each row
df['Column'].map(lambda x: x * 2)    # Map function to each element

```python
#4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]})
corr_matrix = df.corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
#8.  Create a line plot with multiple lines using Matplotlib ?
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]

plt.plot(x, y1, label="Line 1", marker="o")
plt.plot(x, y2, label="Line 2", linestyle="--", marker="s")

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Multiple Line Plot")
plt.legend()
plt.show()

```

```python
#10. Create a histogram using Seaborn to visualize a distribution?
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

data = np.random.randn(1000)

sns.histplot(data, kde=True, bins=30, color='blue')
plt.title("Histogram with KDE")
plt.show()

```
